Common approaches involve filling in missing values in a variable with that variable's **mean**, **median**, or **mode** value.

But some variables contain blank values between 60-70%. Filling these variables with the values we have will not give an accurate result.

*   AVG_MATH_4_SCORE                67.055394
*   AVG_MATH_8_SCORE                64.897959
*   AVG_READING_4_SCORE             62.099125
*   AVG_READING_8_SCORE             67.230321





```python
df['ENROLL'] = df.ENROLL.fillna(df.groupby('YEAR')['ENROLL'].transform('mean'))  
df['ENROLL'] 
```

There are still Nan values because there is no data for some years.
Can be filled with the average of data from recent years.

```python
df['ENROLL'].iloc[90:100]
```

```python
columns=['TOTAL_REVENUE','FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE',
       'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE',
       'SUPPORT_SERVICES_EXPENDITURE', 'OTHER_EXPENDITURE',
       'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G', 'GRADES_KG_G',
       'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G',
       'GRADES_9_12_G', 'GRADES_ALL_G', 'AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE',
       'AVG_READING_4_SCORE', 'AVG_READING_8_SCORE']
```

```python
for column in columns:
    df[column] = df[column].fillna(df.groupby('YEAR')[column].transform('mean'))
```

```python
df.isna().sum()*100/df.shape[0]
```

The basic idea in **interpolation** is to fill in missing values between two non-missing values using these non-missing values.

```python
df3=df.interpolate(method='linear', limit_direction='forward', axis=0)
df3.isna().sum()*100/df.shape[0]
```

As a result, we can say that the best way to fill data in this data set, especially for score variables, is interpolation.
