# 4.Popülasyon, Örneklem ve İlgili Teoremler 

İki normal dağılıma sahip rassal sayı üretim. Bu rassal sayılardan birisinden 1000 tane 10.000 tane elde edelim. Bu iki rassal sayının da olasılık yoğunluk grafiğini çizelim. Sizce Merkezi Limit Theoremi bu olayda geçerli midir?

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

import warnings

warnings.filterwarnings("ignore")
```

```python
mu, sigma, n = 10, 0.2, 100
birinci_rassal_sayi = np.random.normal(mu, sigma, n)

mu, sigma, n = 10, 0.2, 1000
ikinci_rassal_sayi = np.random.normal(mu, sigma, n)

mu, sigma, n = 10, 0.2, 10000
ucuncu_rassal_sayi = np.random.normal(mu, sigma, n)
```

```python
def plot_norm(counts, bins):
    plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi))*np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=3, color='y')
    return None
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
plt.ylim((0,4))
count, bins, ignore = plt.hist(birinci_rassal_sayi, 50, density=True)
plot_norm(count, bins)

plt.subplot(132)
plt.ylim((0,4))
count, bins, ignore = plt.hist(ikinci_rassal_sayi, 50, density=True)
plot_norm(count, bins)

plt.subplot(133)
plt.ylim((0,4))
count, bins, ignore = plt.hist(ucuncu_rassal_sayi, 50, density=True)
plot_norm(count, bins)

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
sns.distplot(birinci_rassal_sayi, bins = 50)

plt.subplot(132)
sns.distplot(ikinci_rassal_sayi, bins = 50)

plt.subplot(133)
sns.distplot(ucuncu_rassal_sayi, bins = 50)

plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
res = stats.probplot(birinci_rassal_sayi, plot=plt)

plt.subplot(132)
res = stats.probplot(ikinci_rassal_sayi, plot=plt)

plt.subplot(133)
res = stats.probplot(ucuncu_rassal_sayi, plot=plt)

plt.show()
```

Merkezi Limit teoremi gecerlidir ancak gozlem sayisi azaldikca etkisi azalmaktadir.

```python
adil_para = np.random.binomial(1,0.5, 100)
yanli_para = np.random.binomial(1,0.75, 100)
print("Adil para : ", sum(adil_para)/len(adil_para))
print("Yanlı para : ", sum(yanli_para)/len(yanli_para))
```
