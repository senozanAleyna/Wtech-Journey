```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format
```

```python
import pandas as pd
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

```python
weather_df.columns
```

```python
weather_df['TGap (C)']=weather_df['Temperature (C)']-weather_df['Apparent Temperature (C)']
```

```python
Y = weather_df['TGap (C)']
X = weather_df[['Humidity','Wind Speed (km/h)']]
X = sm.add_constant(X)
```

```python
sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

Tahmin edilen model: 

$$ TGap (C) = -2.4381 + 3.0292 * Humidity + 0.1193	*Wind Speed (km/h) \qquad(denklem.1)$$ <br>
p-değerleri sıfıra çok yakın olduğundan istatistiksel olarak anlamlıdır.<br>
Humidity 'in modele etkisi Wind Speed'e göre daha fazladır.<br>
Hedef ve açıklayıcı değişken arasında pozitif bir ilişki vardır. <br>
Matematiksel olarak, iki değişkenin etkileşimini, çarpımlarını alarak yakalarız. Şimdi, bir etkileşim değişkeni oluşturalım ve modelimize ekleyelim:

```python
Y = weather_df['TGap (C)']
weather_df['HumidityWind'] = weather_df['Humidity'] * weather_df['Wind Speed (km/h)']
X = weather_df[['Humidity','Wind Speed (km/h)','HumidityWind']]

X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

Tahmin edilen model: 

$$ TGap (C) = 	-0.0839 	-0.1775 * Humidity 	-0.0905	*Wind Speed (km/h) +	0.2971*HumidityWind	 \qquad(denklem.2)$$ <br>
Öncelikle, tüm katsayıların önceki modele göre değişti. <br>
Bunun nedeni, bir değişkeni her eklediğimiz veya çıkarttığımız zaman, model spesifikasyonu değişir. Bu, katsayılar için farklı nokta tahminleri ile sonuçlanır. Aynısı, katsayıların istatistiksel anlamlılığı için de geçerlidir. <br>
Humidity ve Wind Speed modele artık negatif etki yaratmaktadır

**Another Example**

```python
house_prices_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/house_prices.csv", sep = ";")
house_prices_df.head()
```

```python
house_prices_df.columns
```

```python
to_drop = ['Id','Alley','PoolQC','Fence','MiscFeature','Electrical']

for k in to_drop:
    house_prices_df.drop([k], axis = 1, inplace = True)


cat_col_data = ['FireplaceQu','GarageType','GarageFinish','MasVnrType','BsmtQual',
           'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu',
          'GarageQual','GarageCond']

ncat_col_data = ['LotFrontage','GarageYrBlt','MasVnrArea']
for i in cat_col_data:
    house_prices_df[i] = house_prices_df[i].fillna(house_prices_df[i].mode()[0])
    
for j in ncat_col_data:
    house_prices_df[j] = house_prices_df[j].fillna(house_prices_df[j].mean())
```

```python
categorical_features=[feature for feature in house_prices_df.columns if house_prices_df[feature].dtype=='O']
for feature in categorical_features:
    labels_ordered=house_prices_df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    house_prices_df[feature]=house_prices_df[feature].map(labels_ordered)
```

```python
X = house_prices_df.drop(['SalePrice'], axis = 1)
Y = house_prices_df['SalePrice']
```

```python
X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```

P değerleri incelendiğinde;
MSZoning,LotShape,Exterior2nd,Exterior1st,ExterCond,BsmtCond,YrSold,MiscVal,PoolArea,EnclosedPorch,OpenPorchSF,GarageCond,GarageArea,GarageFinish gibi değerler istatiksel olarak anlamsızdır.

```python
X = house_prices_df.drop(['MSZoning','LotShape','Exterior2nd','Exterior1st','ExterCond','BsmtCond','YrSold','MiscVal','PoolArea','EnclosedPorch','OpenPorchSF','GarageCond','GarageArea','GarageFinish','SalePrice'], axis = 1)
Y = house_prices_df['SalePrice']
```

```python
X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```

Anlamsız değişkenleri çıkarmama rağmen henüz bir etki yaratmadı.Korele featurelara bakıp modeli ona göre oluşturacağım.

```python
numericData= house_prices_df.select_dtypes(include=[np.number])
_data = numericData.corr()['SalePrice'][:-1] # -1 because the latest row is SalePrice
golden_features_list = _data[abs(_data) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with SalePrice:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
X = house_prices_df[['OverallQual','GrLivArea','Neighborhood','ExterQual']]
Y = house_prices_df['SalePrice']
X = sm.add_constant(X)
results = sm.OLS(Y, X).fit()
results.summary()
```
