In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd

import matplotlib.pyplot as plt

covid_path = "https://github.com/owid/covid-19-data/raw/master/public/data/owid-covid-data.csv"

df = pd.read_csv(covid_path, parse_dates=["date"], low_memory=False,index_col = 'date')
```

```python
countriesToSelect = ["Spain","France","Germany","Italy"]
df_last_100_day = df.loc['2021-09-05':'2021-12-13']
df_last_100_day.location.isin(countriesToSelect)

df4countries = df_last_100_day[df_last_100_day.location.isin(countriesToSelect)]
df4France = df4countries[:100]
df4Germany = df4countries[100:200]
df4Italy = df4countries[200:300]
df4Spain = df4countries[300:400]
df4Italy

```

```python
plt.figure(figsize=(24,10))

#Total Cases
plt.suptitle('Covid-19 Total Deaths (Last 100 Days)', 
          fontstyle = 'normal', 
          fontsize = 22,
          fontweight = 'normal',
          color = "black",
          fontfamily = 'Arial')

plt.subplot(1, 2, 1)
plt.plot(df4France['total_cases'], label='FRA', c="blue", marker = "o", ms = 2, ls = "-", lw = 1)
plt.plot(df4Germany['total_cases'], label='GER', c="orange", marker = "*", ms = 2, ls = "--", lw = 1)
plt.plot(df4Italy['total_cases'], label='ITA', c="green", marker = "+", ms = 2, ls = ":", lw = 1)
plt.plot(df4Spain['total_cases'], label='ESP', c="red", marker = "v", ms = 2, ls = "-", lw = 1)

plt.title('Total Cases', 
          fontstyle = 'normal',
          fontsize = 18,
          fontweight = 'normal',
          color = "darkred",
          fontfamily = 'Arial')
plt.xlabel('Last 100 days',
          fontstyle = 'normal',
          fontsize = 15,
          fontweight = 'normal',
          color = "darkblue",
          fontfamily = 'Arial')
plt.ylabel('Total Cases (Million)',
          fontstyle = 'normal',
          fontsize = 15,
          fontweight = 'normal',
          color = "darkblue",
          fontfamily = 'Arial')
plt.legend(title = 'Countries', loc = 'lower right')
plt.xticks(rotation = 30, fontsize = 9)

#Total Deaths

plt.subplot(1, 2, 2)
plt.plot(df4France['total_deaths'], label='FRA', c="blue", marker = "o", ms = 2, ls = "-", lw = 1)
plt.plot(df4Germany['total_deaths'], label='GER', c="orange", marker = "*", ms = 2, ls = "--", lw = 1)
plt.plot(df4Italy['total_deaths'], label='ITA', c="green", marker = "+", ms = 2, ls = ":", lw = 1)
plt.plot(df4Spain['total_deaths'], label='ESP', c="red", marker = "v", ms = 2, ls = "-", lw = 1)


plt.title('Total Deaths', 
          fontstyle = 'normal', 
          fontsize = 18,
          fontweight = 'normal',
          color = "darkred",
          fontfamily = 'Arial')

plt.xlabel('Last 100 days',
          fontstyle = 'normal',
          fontsize = 15,
          fontweight = 'normal',
          color = "darkblue",
          fontfamily = 'Arial')
plt.ylabel('Total Deaths',
          fontstyle = 'normal',
          fontsize = 15,
          fontweight = 'normal',
          color = "darkblue",
          fontfamily = 'Arial')
plt.legend(title = 'Countries', loc = 'lower right')
plt.xticks(rotation = 30, fontsize = 9)


plt.show()
```
