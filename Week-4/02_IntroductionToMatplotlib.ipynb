**(1)** Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3$\pi$ and 3$\pi$ with 0.1 intervals.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-3*np.pi,3*np.pi,0.1)
y = np.cos(x)
fig, ax = plt.subplots() # created a figure and a set of subplots.
# Now we have an axes named ax. We can add plots to this axes.
figure=ax.plot(x, y, color = '#22B573')
plt.show()
```

**(2)** This time, create the same plot in red color with object methods. 

```python
fig, ax = plt.subplots() 
ax.plot(x, y, color = 'red')
plt.show()
```

```python
ax.set_xlabel("$x$")
ax.set_ylabel("$\cos(x)$")
ax.set_title("numbers beetwen -3 $\pi$  and 3 $\pi$  with 0.1 intervals")

fig
```
